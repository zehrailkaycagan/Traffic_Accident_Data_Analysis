```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV dosyasının yolu
file_path = './data/accidents_2012_to_2014.csv'

data = pd.read_csv(file_path)

data.head(10)
```

+Accident_Index: Kaza olayının benzersiz kimlik numarası veya indeksi.

+Location_Easting_OSGR: Kaza yerinin Doğu-Batı coğrafi koordinatı (OSGR formatında).

+Location_Northing_OSGR: Kaza yerinin Kuzey-Güney coğrafi koordinatı (OSGR formatında).

+Longitude: Kaza yerinin boylam koordinatı.
+Latitude: Kaza yerinin enlem koordinatı.

+Police_Force: Polis kuvvetinin kimlik numarası veya kodu.

+Accident_Severity: Kaza ciddiyeti (örneğin, ölüm, ciddi yaralanma, hafif yaralanma).

+Number_of_Vehicles: Kaza olayına karışan araçların sayısı.

+Number_of_Casualties: Kaza olayında meydana gelen yaralı veya ölü sayısı.

+Date: Kaza tarihi.

+Pedestrian_Crossing-Physical_Facilities: Yaya geçidi fiziksel özellikleri (örneğin, trafik ışıkları, yaya geçidi).

+Light_Conditions: Işık koşulları (örneğin, gündüz, gece, alacakaranlık).

+Weather_Conditions: Hava koşulları (örneğin, yağmurlu, karlı, güneşli).

+Road_Surface_Conditions: Yol yüzey koşulları (örneğin, kuru, kaygan).

+Special_Conditions_at_Site: Kaza yerindeki özel koşullar (örneğin, inşaat alanı, yol çalışması).

+Carriageway_Hazards: Karayolu üzerindeki tehlikeler (örneğin, kazara dökülmüş yük).

+Urban_or_Rural_Area: Kaza yerinin şehir içi veya kırsal bölge olup olmadığını belirten bir gösterge.

+Did_Police_Officer_Attend_Scene_of_Accident: Kaza yerine polis memuru katılıp katılmadığını belirten bir gösterge.

+LSOA_of_Accident_Location: Kaza yerinin yerel istatistiksel bölge kimliği.

+Year: Kaza olayının meydana geldiği yıl.

```python
import datetime as dt
import warnings
warnings.filterwarnings('ignore')
selRows = data[data['Time'].isna()].index
data = data.drop(selRows, axis=0)
data["Time"] = pd.to_datetime(data["Time"], format='%H:%M')
data["Hour"] = data["Time"].dt.hour
data["Month"] = pd.DatetimeIndex(data["Date"]).month
```

```python
f, axs = plt.subplots(3,2, figsize = (14,14), constrained_layout = True)

road_sayisi = data.groupby("Road_Type")
toplam_road=pd.DataFrame(road_sayisi.size().sort_values(ascending=False),columns=["Road_toplam"])
light_sayisi = data.groupby("Light_Conditions")
toplam_light=pd.DataFrame(light_sayisi.size().sort_values(ascending=False),columns=["Light_toplam"])
weather_sayisi = data.groupby("Weather_Conditions")
toplam_weather=pd.DataFrame(weather_sayisi.size().sort_values(ascending=False),columns=["Weather_toplam"])
surface_sayisi = data.groupby("Road_Surface_Conditions")
toplam_surface=pd.DataFrame(surface_sayisi.size().sort_values(ascending=False),columns=["road_surface_toplam"])
saat_sayisi = data.groupby("Hour")
toplam_saat=pd.DataFrame(saat_sayisi.size().sort_values(ascending=False),columns=["saat_toplam"])
ay_sayisi = data.groupby("Month")
toplam_ay=pd.DataFrame(ay_sayisi.size().sort_values(ascending=False),columns=["ay_toplam"])
                       
                       

ax=sns.barplot(x=toplam_road.index, y=toplam_road["Road_toplam"], data=toplam_road, ax = axs[0, 0])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x=toplam_light.index,y=toplam_light["Light_toplam"], data=toplam_light, ax = axs[0, 1])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x=toplam_weather.index,y=toplam_weather["Weather_toplam"], data=toplam_weather, ax = axs[1, 0])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x=toplam_surface.index,y=toplam_surface["road_surface_toplam"], data=toplam_surface, ax = axs[1, 1])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x=toplam_saat.index,y=toplam_saat["saat_toplam"], data = toplam_saat, ax = axs[2, 0])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x=toplam_ay.index,y=toplam_ay["ay_toplam"], data=toplam_ay, ax = axs[2, 1])
ax.tick_params(axis = 'x', rotation = 45)
axs[0, 0].title.set_text('Yol tipine göre kaza oranları')
axs[0, 1].title.set_text('Işık koşullarına göre kaza oranları')
axs[1, 0].title.set_text('Hava durumuna göre kaza oranları')
axs[1, 1].title.set_text('Yol yüzey koşullarına göre kaza oranları')
axs[2, 0].title.set_text('Saate göre kaza oranları')
axs[2, 1].title.set_text('Aylara göre kaza oranları')
```
